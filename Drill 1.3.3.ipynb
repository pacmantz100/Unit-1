```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
grad_admissions = pd.read_csv('Admission_Predict.csv')
ga=grad_admissions
ga.head()
```

```python
sns.set(style="ticks")
```

```python
g = sns.lmplot(y='GRE Score', # Variable 1.
               x='CGPA', # Variable 2.
               data=ga, # Data
               fit_reg=True, # If set to true, plots a regression line.
               scatter_kws={'alpha':0.4}) # Set points to semi-transparent to see overlaping points.
g.set_ylabels("GRE Score")
g.set_xlabels("CGPA")
plt.title('Scatterplot')
plt.show()
```

```python
# Setting the overall aesthetic.
sns.set(style="white")

f = sns.lmplot(y='GRE Score', 
               x='TOEFL Score',
               hue='University Rating',
               data=ga, # Data.
               col='Research',
               fit_reg=True,
               ci=True,
               scatter_kws={'alpha':0.4})
f.set_ylabels("GRE Score")
f.set_xlabels("TOEFL Score")
plt.suptitle('Scatterplot Regression')
plt.subplots_adjust(top=0.9)
plt.show()
```
